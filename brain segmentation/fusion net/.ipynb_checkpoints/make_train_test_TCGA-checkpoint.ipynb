```python
import os
import numpy as np
import nibabel as nib
import scipy.misc
```

## Make Train, Test for TCGA data

#### TCGA data : 

```python
def nii2numpy(nii):
    return nib.load(nii).get_data()

def save_axial_image(img, subject, path):
    dt=0
    try:
        w, h, n=img.shape
    except:
        print('check ',subject)
        return None
    for x in range(n):
        filename=subject+'_'+str(dt)+'.jpg'
        scipy.misc.imsave(os.path.join(path,subject,filename), img[:,:,x])
        dt+=1
                
def make_ref_imgs(masks, whole_path):
    '''making_input_image_data'''
    image_paths = []
    mask_paths = []

    for mask in masks:
        subject = mask[:7]
        try:
            image_names = whole_path[subject]
        except:
            continue
#         mask_paths.append(os.path.join(folder_path, mask_path, mask))
        mask_paths.append(mask)

        if 'F' in mask:
            for name in image_names:
                if 'F' in name and 'mask' not in name:
#                     image_paths.append(os.path.join(folder_path, origin_path, subject, name))
                    image_paths.append(name)
        elif '5588560' in mask:
            for name in image_names:
                if 'F' in name and '5588560' in name:
                    image_paths.append(name)
        elif '7393641' in mask:
            for name in image_names:
                if 'F' in name and '7393641' in name:
                    image_paths.append(name)
        else:
            for name in image_names:
                if 'T2' in name and 'F' not in name and 'mask' not in name:
#                     image_paths.append(os.path.join(folder_path, origin_path, subject, name))
                    image_paths.append(name)
    return image_paths, mask_paths

def get_ref_img(subject):
    for img in image_paths:
        if subject in img:
            return img
```

```python
'''type data folder path'''
folder_path = '/base/folder/path'
save_original_path='/save/original/path'
save_mask_path='/save/mask/path'

subjects=os.listdir(folder_path)
whole_path={}
'''Get one subject whole data'''

for subject in subjects:
    imgs=os.listdir(os.path.join(folder_path, subject))
    for img in imgs:
        if 'mask' in img:
            mask_img=os.path.join(folder_path, subject, img)
        if 'T2' in img and 'FLAIR' not in img:
            origin_img=os.path.join(folder_path, subject, img)
            
    whole_path[subject]=[origin_img, mask_img]
```

```python
for subject in subjects:
    if subject not in ['CS_4941','CS_5395', 'CS_5397', 'CS_6186','CS_6188' ,'CS_6669', 'DU_5852']:
        origin_img, mask_img=whole_path[subject]

        try:
            os.mkdir(os.path.join(save_original_path, subject))
            os.mkdir(os.path.join(save_mask_path, subject))
        except:
            pass

        origin_data=nib.load(origin_img).get_data()
        mask_data=nib.load(mask_img).get_data()

        save_axial_image(origin_data, subject, save_original_path)
        save_axial_image(mask_data, subject, save_mask_path)

        print('{} done'.format(subject))
```

```python
# Train data
for im in image_paths:
    s=im[:7]
    img=nii2numpy(os.path.join(folder_path, origin_path, s, im))
    os.mkdir('./data/train/{}'.format(s))
    save_train(img,s)
    print(s,'--- Done')
```

```python
# Train mask data
error=[]
for im in mask_paths:
    s=im[:7]
    try:
        print(s,' is now processed')
        img=nii2numpy(os.path.join(folder_path, mask_path, im))
        os.mkdir('../../segmentation/Pytorch-UNet/data/train_masks/{}'.format(s))
        save_mask(img,s)
    except:
        error.append(s)
        pass
```
